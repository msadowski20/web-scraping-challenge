```python
# Dependencies
from bs4 import BeautifulSoup
import requests
from splinter import Browser
import pandas as pd
```

```python
news_url = 'https://mars.nasa.gov/news/'
response = requests.get(news_url)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
results = soup.find_all('div', class_='slide')
results
```

```python
title_list = []
p_list = []
```

```python
for result in results:
    title = result.find('div', class_="content_title").text
    p = result.div.a.text
    
    title_list.append(title)
    p_list.append(p)

    print('-------------')
    print(title)
    print(p)
```

```python
news_title = title_list[0].strip()
news_p = p_list[0].strip()

news_title
```

```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
hemisphere_url = 'https://astrogeology.usgs.gov/search/results?q=hemisphere+enhanced&k1=target&v1=Mars'
browser.visit(hemisphere_url)
```

```python
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
```

```python
results = soup.find_all('div', class_='item')
results
```

```python
hemisphere_list = []
url_list = []

for result in results:
    hemisphere = result.h3.text
    hemisphere_name = hemisphere[:-9]
    hemisphere_list.append(hemisphere_name)
    
    browser.click_link_by_partial_text(hemisphere)
    html = browser.html
    soup = BeautifulSoup(html, 'html.parser')
    image_link = soup.find('a', class_='href').text
    url_list.append(image_link)
    browser.back()
    
url_list
```
