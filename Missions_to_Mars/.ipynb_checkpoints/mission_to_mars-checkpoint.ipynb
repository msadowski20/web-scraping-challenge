```python
# Dependencies
from bs4 import BeautifulSoup
import requests
from splinter import Browser
import pandas as pd
```

```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# Set url and create soup for Mars news site
news_url = 'https://mars.nasa.gov/news/'
browser.visit(news_url)
```

```python
# Create soup from news link html
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
```

```python
# Scrape Mars news site to find article reference
results = soup.find_all('div', class_='slide')
results
```

```python
title_list = []
p_list = []
```

```python
# Loop through results to get title and paragraph for each article
for result in results:
    title = result.find('div', class_="content_title").text
    p = result.div.a.text
    
    title_list.append(title)
    p_list.append(p)

    print('-------------')
    print(title)
    print(p)
```

```python
# SAVE VARIABLES: Extract the first entry for title and paragraph
news_title = title_list[0].strip()
news_p = p_list[0].strip()

news_p
```

```python
# Set twitter link and create soup
twitter_url = 'https://twitter.com/marswxreport?lang=en'
response = requests.get(twitter_url)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
# Scrape twitter site to find the most current tweet with the weather
results = soup.find_all('div', class_='tweet')
results
```

```python
# Loop through scrape results and append weather tweets to a list
weather_list = []

for result in results:
    tweet = result.find('div', class_="js-tweet-text-container")
    weather = tweet.p.text
    weather_list.append(weather)
    
weather_list
```

```python
# SAVE VARIABLES: Extract most current tweet and format string
mars_weather = weather_list[0].strip()
mars_weather
```

```python
# Initialize chrome driver
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# Set Mars astrology links for chrome driver
hemisphere_url = 'https://astrogeology.usgs.gov/search/results?q=hemisphere+enhanced&k1=target&v1=Mars'
browser.visit(hemisphere_url)
```

```python
# Create soup from Mars astrology link html
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
```

```python
# Scrape site to find the hemisphere containers
results = soup.find_all('div', class_='item')
results
```

```python
# Loop through the results to get the title of each link
# use splinter TO click on each hemisphere link and save the image url
hemisphere_base_url = 'https://astrogeology.usgs.gov'
image_url_dict = []

for result in results:
    title = result.h3.text
    partial_image_url = result.find('a', class_='itemLink product-item')['href']
    browser.visit(hemisphere_base_url + partial_image_url)
    
    image_html = browser.html
    soup = BeautifulSoup(image_html, 'html.parser')
    image_url = hemisphere_base_url + soup.find('img', class_='wide-image')['src']
    
    image_url_dict.append({"title" : title, "image_url" : image_url})
    

    
```

```python
# save VARIABLES: Save the title and image url to a dictionary
image_url_dict
```
